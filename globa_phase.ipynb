```python
import qiskit, qiskit.visualization as qv
import numpy as np
```

```python
qc = qiskit.QuantumCircuit(2)
qc.h(0)
qc.h(1)
qc.z(1)
```

```python
qv.plot_bloch_multivector(qc)
```

```python
import qiskit.quantum_info


qc2 = qiskit.QuantumCircuit(1)
a =qiskit.quantum_info.Statevector([np.sqrt(3)/2,1/2])
qc2.initialize(a,0)
```

```python
qc2.y(0)
qv.plot_bloch_multivector(qc2)
```
