```python
import qiskit
```

```python
qc = qiskit.QuantumCircuit(2)
qc.h(0)
qc.cx(0,1)
qc.draw('mpl')
```

```python
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
pm = generate_preset_pass_manager(optimization_level=3, basis_gates=['x','sx','rz','id','ecr'])
trans_qc = pm.run(qc)
trans_qc.draw('mpl')
```
