```python
import qiskit
import qiskit.visualization as qv
```

```python
qc = qiskit.QuantumCircuit(3,3)

# Create a random qubit
qc.h(0)

qc.barrier()

#  Create a bell state
qc.h(1)
qc.cx(1,2)

qc.barrier()

# Teleportation
qc.cx(0,1)
qc.h(0)

qc.barrier()

#  Measuring
qc.measure(0,0)
qc.measure(1,1)

qc.barrier()

# Telephone call
qc.cx(1,2)
qc.cz(0,2)
qc.measure(2,2)
```

```python
qc.draw("mpl", initial_state=True)
```

```python
from qiskit.providers.basic_provider import BasicSimulator
backend = BasicSimulator()
tqc = qiskit.transpile(qc, backend)
counts = backend.run(tqc).result().get_counts()
qv.plot_histogram(counts)
```
