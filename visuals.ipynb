```python
import qiskit
import qiskit.visualization as qv
```

```python
qc = qiskit.QuantumCircuit(2)
qc.h(0)
qc.x(1)
```

```python
qc.draw("mpl")
```

```python
statevector = qiskit.quantum_info.Statevector(qc)
statevector.draw('latex')
```

```python
qv.plot_state_qsphere(qc)
```

```python
qv.plot_bloch_multivector(qc)
```
